```python
# Очистка данных

# Проверка и удаление ненужных столбцов, если они существуют
columns_to_drop = ['Scrape ID', 'Last Scraped', 'Experiences Offered', 'Thumbnail Url', 'Medium Url', 'XL Picture Url']
existing_columns = [col for col in columns_to_drop if col in df.columns]
df = df.drop(columns=existing_columns)

# Обработка пропущенных значений и преобразование типов данных
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df['Review Scores Rating'] = pd.to_numeric(df['Review Scores Rating'], errors='coerce')
df['Review Scores Rating'] = df['Review Scores Rating'].fillna(df['Review Scores Rating'].mean())

# Преобразование дат
df['Host Since'] = pd.to_datetime(df['Host Since'], errors='coerce')
df['Last Review'] = pd.to_datetime(df['Last Review'], errors='coerce')

# Проверка типов данных
print(df.dtypes)

# Базовая статистика
print(df[['Price', 'Review Scores Rating']].describe())

# Вывод первых нескольких строк для проверки
print(df.head())
```

Почитаем статистику по ценам и рейтингу

```python
import pandas as pd

# Загрузка данных
df = pd.read_csv('airbnb-listings.csv', sep=';', encoding='utf-8', low_memory=False)

# Анализ цен
average_price = df['Price'].mean()
median_price = df['Price'].median()
min_price = df['Price'].min()
max_price = df['Price'].max()
price_percentiles = df['Price'].quantile([0.25, 0.5, 0.75])

print("Анализ цен:")
print(f"Средняя цена: ${average_price:.2f}")
print(f"Медиана цены: ${median_price:.2f}")
print(f"Минимальная цена: ${min_price:.2f}")
print(f"Максимальная цена: ${max_price:.2f}")
print("Распределение цен по процентам:")
print(price_percentiles)

# Анализ рейтингов
average_rating = df['Review Scores Rating'].mean()
min_rating = df['Review Scores Rating'].min()
max_rating = df['Review Scores Rating'].max()
rating_percentiles = df['Review Scores Rating'].quantile([0.25, 0.5, 0.75])

print("\nАнализ рейтингов:")
print(f"Средний рейтинг: {average_rating:.2f}")
print(f"Минимальный рейтинг: {min_rating:.2f}")
print(f"Максимальный рейтинг: {max_rating:.2f}")
print("50% объявлений в диапазоне рейтинга:")
print(f"От {rating_percentiles[0.25]:.2f} до {rating_percentiles[0.75]:.2f}")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import folium
import random

# Данные для графика
countries = ['США', 'Россия', 'Германия', 'Италия', 'Испания', 'ЮАР', 'Индонезия']
listings = [1000] * 7
prices = [236.95, 86, 100, 129, 132, 50, 39]
occupancy = [73, 54, 73, 84, 86, 60, 49]


# Создание графика
x = np.arange(len(countries))
width = 0.35

fig, ax1 = plt.subplots(figsize=(12, 6))
ax2 = ax1.twinx()

rects1 = ax1.bar(x - width/2, prices, width, label='Средняя цена за ночь ($)', color='skyblue')
rects2 = ax2.bar(x + width/2, occupancy, width, label='Средняя заполняемость (%)', color='lightgreen')

ax1.set_xlabel('Страны')
ax1.set_ylabel('Средняя цена за ночь ($)')
ax2.set_ylabel('Средняя заполняемость (%)')
ax1.set_title('Сравнение цен и заполняемости Airbnb по странам')
ax1.set_xticks(x)
ax1.set_xticklabels(countries)

ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

fig.tight_layout()
plt.savefig('airbnb_comparison.png')
plt.close()

# Создание карты
m = folium.Map(location=[0, 0], zoom_start=2)

# Примерные координаты центров стран
country_coords = {
    'США': [37.0902, -95.7129],
    'Россия': [61.5240, 105.3188],
    'Германия': [51.1657, 10.4515],
    'Италия': [41.8719, 12.5674],
    'Испания': [40.4637, -3.7492],
    'ЮАР': [-30.5595, 22.9375],
    'Индонезия': [-0.7893, 113.9213]
}

for country, coords in country_coords.items():
    # Добавляем маркер для каждой страны
    folium.Marker(
        location=coords,
        popup=f"{country}<br>Средняя цена: ${prices[countries.index(country)]}<br>Заполняемость: {occupancy[countries.index(country)]}%",
        tooltip=country
    ).add_to(m)

    # Добавляем 100 случайных точек вокруг центра страны
    for _ in range(1000):
        lat = coords[0] + random.uniform(-5, 5)
        lon = coords[1] + random.uniform(-5, 5)
        folium.CircleMarker(
            location=[lat, lon],
            radius=2,
            popup=f"{country}<br>Цена: ${prices[countries.index(country)]}",
            color='red',
            fill=True
        ).add_to(m)

# Сохраняем карту
m.save("airbnb_map_new.html")

print("График сохранен как 'airbnb_comparison.png', а карта - как 'airbnb_map_new.html'")
```

```python
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd

# Загрузка данных
df = pd.read_csv('airbnb-listings.csv', sep=';', encoding='utf-8', low_memory=False)

# Инициализация приложения Dash
app = dash.Dash(__name__)

# Определение макета дашборда
app.layout = html.Div([
    html.H1("Анализ данных Airbnb"),

    dcc.Dropdown(
        id='country-dropdown',
        options=[{'label': i, 'value': i} for i in df['Country'].unique()],
        value='United States',
        multi=False,
        style={'width': '50%'}
    ),

    dcc.Graph(id='price-distribution'),

    dcc.Graph(id='room-type-distribution'),

    dcc.Graph(id='price-vs-rating')
])

# Колбэк для обновления графиков при выборе страны
@app.callback(
    [Output('price-distribution', 'figure'),
     Output('room-type-distribution', 'figure'),
     Output('price-vs-rating', 'figure')],
    [Input('country-dropdown', 'value')]
)
def update_graphs(selected_country):
    filtered_df = df[df['Country'] == selected_country]

    # График распределения цен
    price_fig = px.histogram(filtered_df, x='Price', nbins=50,
                             title=f'Распределение цен в {selected_country}')

    # График распределения типов комнат
    room_type_fig = px.pie(filtered_df, names='Room Type',
                           title=f'Распределение типов комнат в {selected_country}')

    # График зависимости цены от рейтинга
    price_vs_rating_fig = px.scatter(filtered_df, x='Review Scores Rating', y='Price',
                                     title=f'Цена vs Рейтинг в {selected_country}')

    return price_fig, room_type_fig, price_vs_rating_fig

# Запуск приложения
if __name__ == '__main__':
    app.run_server(debug=True)
```

### Анализ сезонности

```python
import pandas as pd
import matplotlib.pyplot as plt

# Загрузка данных
df = pd.read_csv('airbnb-listings.csv', sep=';', encoding='utf-8', low_memory=False)

# Убедимся, что даты в правильном формате
df['Last Review'] = pd.to_datetime(df['Last Review'], errors='coerce')

# Добавляем столбец с месяцами
df['Month'] = df['Last Review'].dt.month

# Рассчитываем среднюю цену за каждый месяц
monthly_prices = df.groupby('Month')['Price'].mean()

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(monthly_prices.index, monthly_prices.values, marker='o')
plt.title('Сезонность цен на Airbnb')
plt.xlabel('Месяц')
plt.ylabel('Средняя цена за ночь ($)')
plt.grid(True)
plt.xticks(range(1, 13))
plt.show()
```

По данному графику анализа сезонности очевидно, что пик цен кварир прихожится на 5 - 7 месяца, то есть на май - июль

Сделаем для наглядности и тренировки также график Pyechart

```python
import pandas as pd
from pyecharts import options as opts
from pyecharts.charts import Map, Page
from pyecharts.globals import ThemeType

# Загрузка данных
df = pd.read_csv('airbnb-listings.csv', sep=';', encoding='utf-8', low_memory=False)

# Расширенный список стран
countries = [
    'United States', 'Russia', 'Germany', 'Italy', 'Spain', 'South Africa', 'Indonesia',
    # Дополнительные страны Африки
    'Egypt', 'Kenya', 'Morocco', 'Nigeria', 'Tanzania', 'Ethiopia', 'Ghana', 'Uganda',
    'Algeria', 'Tunisia', 'Senegal', 'Ivory Coast', 'Cameroon', 'Zimbabwe', 'Zambia',
    # Дополнительные страны Азии
    'China', 'Japan', 'India', 'Thailand', 'Vietnam', 'Malaysia', 'Singapore', 'South Korea',
    'Philippines', 'Pakistan', 'Bangladesh', 'Sri Lanka', 'Nepal', 'Cambodia', 'Myanmar',
    # Дополнительные страны Европы
    'France', 'United Kingdom', 'Netherlands', 'Sweden', 'Norway', 'Denmark', 'Finland',
    'Poland', 'Czech Republic', 'Portugal', 'Austria', 'Switzerland', 'Belgium', 'Ireland',
    'Greece', 'Hungary', 'Romania', 'Bulgaria', 'Croatia', 'Serbia',
    # Другие крупные страны
    'Brazil', 'Canada', 'Australia', 'New Zealand', 'Mexico', 'Argentina', 'Chile', 'Colombia',
    'Peru', 'Venezuela', 'Turkey', 'Iran', 'Saudi Arabia', 'United Arab Emirates', 'Israel'
]

# Расчет средних цен по странам
country_prices = df[df['Country'].isin(countries)].groupby('Country')['Price'].mean().round(2)

# Создание карты
world_map = (
    Map(init_opts=opts.InitOpts(width="1200px", height="600px", theme=ThemeType.WESTEROS))
    .add("Средняя цена Airbnb", [list(z) for z in zip(country_prices.index, country_prices.values)], "world")
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
    .set_global_opts(
        title_opts=opts.TitleOpts(title="Средние цены Airbnb по странам"),
        visualmap_opts=opts.VisualMapOpts(
            max_=300,
            is_piecewise=True,
            pieces=[
                {"min": 0, "max": 50, "label": "0-50", "color": "#FFE1E1"},
                {"min": 51, "max": 100, "label": "51-100", "color": "#FFA07A"},
                {"min": 101, "max": 150, "label": "101-150", "color": "#FF7F50"},
                {"min": 151, "max": 200, "label": "151-200", "color": "#FF4500"},
                {"min": 201, "max": 300, "label": "201-300", "color": "#8B0000"}
            ]
        ),
        tooltip_opts=opts.TooltipOpts(formatter="{b}: ${c}")
    )
)

# Создание страницы и добавление карты
page = Page(layout=Page.DraggablePageLayout)
page.add(world_map)

# Сохранение результата в HTML файл
page.render("airbnb_world_prices_extended.html")

print("График сохранен в файл 'airbnb_world_prices_extended.html'")

# Вывод средних цен для проверки
print(country_prices)
```
